```python
import numpy as np
import csv

# 加载数据集
dataset = np.load("datasets/Dataset.npy")

# 输出文件名
output_file = "user_rating_temp.csv"

# 打开文件以写入模式创建 csv.writer 对象
with open(output_file, mode='w', newline='') as file:
    writer = csv.writer(file)
    
    # 写入 header
    writer.writerow(["userID", "tconst", "rating", "reviewDate"])
    
    # 写入数据
    for record in dataset:
        # 将每个记录按照逗号分隔符拆分成字段
        fields = record.split(',')
        # 写入拆分后的字段到 CSV 文件
        writer.writerow(fields)

# 打印提示信息
print(f"Data has been written to {output_file}")
```
