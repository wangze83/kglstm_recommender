```python
data = pd.read_csv('movie_ratings_data_filled.csv')
```

```python
# 将directors, cast, genres列转换为字符串

# 定义一个函数来处理各列的数据
def process_column(column_data):
    try:
        # 尝试用eval处理，如果不成功，则直接返回原始数据
        return ','.join(eval(column_data))
    except:
        # 如果eval失败，直接返回原始数据
        return column_data

data['genres'] = data['genres'].apply(process_column)
data['directors'] = data['directors'].apply(process_column)
data['cast'] = data['cast'].apply(process_column)


```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. 用户行为分析
# 统计每位用户的评分次数分布
user_counts = data['user_id'].value_counts()
plt.figure(figsize=(10, 6))
sns.histplot(user_counts, bins=50, kde=False)
plt.title('Distribution of Ratings Count per User')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Users')
plt.show()
```

```python
# 查看用户评分的平均分布情况
user_avg_rating = data.groupby('user_id')['rating'].mean()
plt.figure(figsize=(10, 6))
sns.histplot(user_avg_rating, bins=50, kde=False)
plt.title('Distribution of Average Ratings per User')
plt.xlabel('Average Rating')
plt.ylabel('Number of Users')
plt.show()
```

```python
# 2. 电影评分分析
# 统计电影的平均评分分布
movie_avg_rating = data.groupby('tconst')['rating'].mean()
plt.figure(figsize=(10, 6))
sns.histplot(movie_avg_rating, bins=50, kde=False)
plt.title('Distribution of Average Ratings per Movie')
plt.xlabel('Average Rating')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# 查看电影的评分数量分布
movie_num_ratings = data.groupby('tconst').size()
plt.figure(figsize=(10, 6))
sns.histplot(movie_num_ratings, bins=50, kde=False)
plt.title('Distribution of Number of Ratings per Movie')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Movies')
plt.show()
```

```python
# 3. 特征相关性分析
# 绘制相关性矩阵图
plt.figure(figsize=(10, 8))
corr_matrix = data[['year', 'averageRating', 'numVotes', 'rating']].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
# 4. 类别分析
# 统计电影类别（genres）的分布情况
genres_count = data['genres'].str.split(',').explode().value_counts()
plt.figure(figsize=(12, 6))
genres_count.plot(kind='bar')
plt.title('Distribution of Movie Genres')
plt.xlabel('Genres')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

```python
# 5. 时间序列分析
# 将 review_date 转换为 datetime 格式
data['review_date'] = pd.to_datetime(data['review_date'])

# 按年份统计评分数量变化
rating_counts_yearly = data.resample('Y', on='review_date')['rating'].count()
plt.figure(figsize=(12, 6))
rating_counts_yearly.plot(marker='o')
plt.title('Number of Ratings Over Time (Yearly)')
plt.xlabel('Year')
plt.ylabel('Number of Ratings')
plt.show()
```

```python
# 6. 用户和电影的交互分析
# 统计每个用户评分的电影数量分布
user_num_ratings = data.groupby('user_id').size()
plt.figure(figsize=(10, 6))
sns.histplot(user_num_ratings, bins=50, kde=False)
plt.title('Distribution of Number of Ratings per User')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Users')
plt.show()
```

```python
# 7. 缺失值分析（如果适用）
# 检查是否存在缺失值
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
# 计算每个评分的数量
rating_distribution = df['rating'].value_counts().sort_index()

# 绘制分布图
plt.figure(figsize=(10, 6))
rating_distribution.plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')
plt.show()

# 检查样本是否均衡
print(rating_distribution)
```

```python
# 计算每个用户评分的数量
user_rating_counts = df['user_id'].value_counts()

# 绘制分布图
plt.figure(figsize=(10, 6))
plt.hist(user_rating_counts, bins=50, edgecolor='k')
plt.xlabel('Number of Ratings per User')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Ratings per User')
plt.show()

# 查找少数用户对大部分电影进行评分的情况
top_users = user_rating_counts[user_rating_counts > user_rating_counts.mean() + 3 * user_rating_counts.std()]
print(f'Number of top rating users: {len(top_users)}')
```

```python
# 计算每个用户评分的数量
user_rating_counts = df['user_id'].value_counts()

# 输出计算结果详情
print(user_rating_counts.describe())
print(user_rating_counts.head())
print(user_rating_counts.tail())

# 查找少数用户对大部分电影进行评分的情况
top_users = user_rating_counts[user_rating_counts > user_rating_counts.mean() + 3 * user_rating_counts.std()]
print(f'Number of top rating users: {len(top_users)}')
print(top_users)
```

```python
# 设置最大评分数阈值
max_ratings_threshold = 500

# 对高频率评分用户进行抽样
def sample_user_ratings(user_df, max_ratings_threshold):
    if len(user_df) > max_ratings_threshold:
        return user_df.sample(max_ratings_threshold)
    else:
        return user_df

balanced_df = df.groupby('user_id').apply(sample_user_ratings, max_ratings_threshold).reset_index(drop=True)

# 检查新数据集的分布
new_user_rating_counts = balanced_df['user_id'].value_counts()
print(new_user_rating_counts.describe())
plt.figure(figsize=(10, 6))
plt.hist(new_user_rating_counts, bins=50, edgecolor='k')
plt.xlabel('Number of Ratings per User')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Ratings per User (After Balancing)')
plt.show()
```

```python
# 计算每部电影的评分数量
movie_ratings_count = df.groupby('tconst')['rating'].count()

# 设置电影评分数量阈值
max_movie_ratings_threshold = 500

# 对评分较多的电影进行抽样
def sample_movie_ratings(movie_df, max_movie_ratings_threshold):
    if len(movie_df) > max_movie_ratings_threshold:
        return movie_df.sample(max_movie_ratings_threshold)
    else:
        return movie_df

balanced_df = balanced_df.groupby('tconst').apply(sample_movie_ratings, max_movie_ratings_threshold).reset_index(drop=True)

# 检查新数据集的分布
new_movie_rating_counts = balanced_df['tconst'].value_counts()
print(new_movie_rating_counts.describe())
plt.figure(figsize=(10, 6))
plt.hist(new_movie_rating_counts, bins=50, edgecolor='k')
plt.xlabel('Number of Ratings per Movie')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Ratings per Movie (After Balancing)')
plt.show()
```
